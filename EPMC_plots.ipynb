```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import pylab as pyl
import math
```

```python
data = pd.read_csv('/path/to/SRM_master.csv')
```

```python
data.sort_values(by='term_score',ascending =True)
```

**EPMC PLOTS**

```python
data = data.sort_values(by='EPMC%coverage', ascending = False)
data.plot.line(x ='term', y='EPMC%coverage')
```

```python
score_term = data['term_score'] > 5

data = data[score_term]
data = data.sort_values(by='EPMC%coverage', ascending = False)
X = data['EPMC%coverage']
Y1 = data['term_score']
Y2 = data['idf_score']*13000

plt.scatter(X,Y1,marker='.')
plt.scatter(X,Y2,marker='.')
plt.legend()
plt.xlabel('EPMC%coverage')
plt.show()
```

```python
data = data.sort_values(by='term_score', ascending = False)

plt.plot(data['term_score'].values, data['%coverage'].values)

# Label the axes
plt.xlabel('term_score')
plt.ylabel('%coverage')

plt.axis([0,130000,0,100])

plt.show()
```

```python
np.corrcoef(data['term_score'],data['EPMC%coverage'])
```

```python
data = data.sort_values(by='idf_score', ascending = False)[11:]
plt.plot(data['idf_score'].values, data['EPMC%coverage'].values)

# Label the axes
plt.xlabel('idf_score')
plt.ylabel('%coverage')

plt.axis([1,11.1,0,100])

plt.show()
```

```python
np.corrcoef(data['idf_score'],data['EPMC%coverage'])
```
